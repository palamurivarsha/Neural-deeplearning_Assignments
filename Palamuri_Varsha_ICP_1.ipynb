```python
x =input("Enter a string:") # Provide the string as input to a variable name varsha
y=list(x) # This will convert to list of characters to use it for deletion
del y[:2] # As asked We've to delete first 2 characters 
output= ''.join(y[::-1]) # this will reverse the string using slicing & join converts the list back to a string
print(output)

'''Time Complexity: O(n)
Space Complexity: O(n)'''
```

```python
x=float(input("Enter first number:")) # Take first number as input from user
y=float(input("Enter second number:")) # Take second number as input from user
print(x+y)# perform addition operation and print output
print(x-y)# perform substraction operation and print output
print(x*y) #  perform multiplication operation and print output
if y!=0: # checking whether second number is '0' because to avoid runtime error
    print(x/y)
else:
    print("Division is not possible")
    
'''
Time Complexity: O(1)
Space Complexity: O(1)'''
```

```python
x=input("Enter a sentence:") # take a sentence as an input from user
y=x.replace("python","pythons") # this line will replace all occurrences of 'python' to 'pythons' using the built-in string method called 'replce()' 
print(y)

'''
Time Complexity: O(n)

Space Complexity: O(n)'''
```

```python
x=float(input("Enter class score:")) # take input from the user

if x>= 90:
    grade="A"
elif x >=80:
    grade="B"
elif x >=70:
    grade="C"
elif x >=60:
    grade="D"
else:
    grade="F"
print("Grade:",grade)

'''
Time Complexity: O(1)

Space Complexity: O(1)'''
```

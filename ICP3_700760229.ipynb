```python
# Used to read the all columns in the given file
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

```python
# used to view all the contents in the file
print(df.describe())
```

```python
# print null values
print(df.isnull().sum())
```

```python
# Replaces missing values with column mean , numeric_only=True → only numeric columns & inplace=True, modifies original data
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# User selects columns, agg() → performs multiple operations at once
c1 = input("Enter first column: ")
c2 = input("Enter second column: ")

print(df[[c1, c2]].agg(['min','max','count','mean']))
```

```python
# Boolean conditions filter rows , & → AND operator
cal = input("Enter Calories column name: ")
print(df[(df[cal] >= 500) & (df[cal] <= 1000)])
```

```python
# Filters rows based on two conditions

pulse = input("Enter Pulse column name: ")
print(df[(df[cal] > 500) & (df[pulse] < 100)])
```

```python
# drop the column from the csv file but not from data frame
remove_col = input("Enter column to remove: ")
df_modified = df.drop(columns=[remove_col])
```

```python
# Delete the column permanentely from data frame
df.drop(columns=[remove_col], inplace=True)
```

```python
# Converts datatype to integer , Useful for calculations and storage
df[cal] = df[cal].astype(int)
print(df.dtypes)
```

```python
# Scatter plot shows relationship, Useful for trend analysis
import matplotlib.pyplot as plt

dur = input("Enter Duration column: ")
df.plot(kind='scatter', x=dur, y=cal)
plt.show()
```

```python
# Reads salary dataset & Shows sample data
import pandas as pd

df = pd.read_csv("Salary_Data (3).csv")
df.head()
```

```python
# ⅔ training, ⅓ testing & Prevents overfitting
from sklearn.model_selection import train_test_split

x_col = input("Enter input column: ")
y_col = input("Enter output column: ")

X = df[[x_col]]
y = df[y_col]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)
```

```python
# Model learns relationship & Predicts salary for test data
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Measures prediction error & Lower MSE = better model
from sklearn.metrics import mean_squared_error

print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train))
plt.show()
#Dots → actual data & Line → regression line
```
